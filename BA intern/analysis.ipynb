
## Task 2 : Gender Distribution

_Objective: Visualize gender distribution in
 the dataset_

```python
gender_counts = data['gender'].value_counts()
print(gender_counts)
```

```python
# Building a Bar chart :
gender_counts.plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Building a Pie chart
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'pink'])
plt.title('Gender Distribution')
plt.ylabel('')  
plt.show()
```

**_There is a notable gender disparity in the dataset, with males (62.5%) outnumbering females (37.5%), indicating a male-dominated sample_**

## Task 3 : Descriptive Statistics
_Objective: Present basic statistics for numerical columns_

```python
numerical_columns = data.select_dtypes(include=['number']).columns
print("Numerical Columns:", numerical_columns)
```

```python
# Calculating basic statistics
stats = {
    'mean': data[numerical_columns].mean(),
    'median': data[numerical_columns].median(),
    'std': data[numerical_columns].std()
}

# Combining statistics into a DataFrame
stats_df = pd.DataFrame(stats)
print(stats_df)
```

### Basic analysis till now : 
_The above statistics reveals that the majority of individuals in the dataset are young, with an average age of 27.8 years. Among the investment avenues, **Gold** (mean: 5.975) and **Debentures** (mean: 5.75) are the most favored options, indicating a preference for secure and tangible investments. Conversely, **PPF (Public Provident Fund)** is the least preferred (mean: 2.025), suggesting limited interest in long-term government savings schemes. The variability is relatively low across most columns, indicating consistent investment preferences within the group_

## Task 4 : Most Preferred Investment Avenue
_Objective: Identify the most preferred investment avenue_

```python
# Sum each investment avenue to find total preference scores
investment_totals = data[['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']].sum()

# Find the most preferred investment avenue
most_preferred = investment_totals.idxmax()
print(f"Most Preferred Investment Avenue: {most_preferred} (Total: {investment_totals[most_preferred]})")
```

## Task 5: Reasons for Investment
_Objective: Analyze and summarize reasons for investment choices_

```python
reason_columns = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']
reasons_data = data[reason_columns].melt(var_name='Investment_Type', value_name='Reason')
reason_counts = reasons_data['Reason'].value_counts()

plt.figure(figsize=(10, 6))
reason_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Reasons for Investment Choices', fontsize=16)
plt.xlabel('Reason', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Summary : 
The chart shows that "Capital Appreciation," "Assured Returns," and "Better Returns" are the top reasons for investment. Other factors like "Risk-Free" and "Safe Investment" are moderately important, while tax and liquidity considerations are less prioritized.

## Task 6: Savings Objectives

_Objective: Identify and present main savings objectives_

```python
saving_Objective = data['What are your savings objectives?'].value_counts()

plt.figure(figsize=(8, 5))
saving_Objective.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Saving objective', fontsize=16)
plt.xlabel('saving_objective', fontsize=14)
plt.xticks(rotation=0, fontsize=10)
```

### _Saving objective summary_:

The primary savings objective among participants is **Retirement Planning**, accounting for the majority (24 responses). This indicates a strong focus on long-term financial security. **Health Care** ranks second with 13 responses, reflecting concern for medical expenses and unforeseen health-related costs. Lastly, **Education** is mentioned by 3 participants, highlighting a smaller yet important focus on funding educational goals. 

This distribution suggests that most participants prioritize stability and preparedness for life’s major events over other potential objectives.

## Task 7: Common Information Sources
_Objective: Analyze common sources participants rely on for investment information_

```python
info_sources = data['Source'].value_counts()
print(info_sources)
```

```python
info_sources = {
    "Financial Consultants": 16,
    "Newspapers and Magazines": 14,
    "Television": 6,
    "Internet": 4
}

labels = info_sources.keys()
sizes = info_sources.values()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'orange', 'pink'])
plt.title("Common Sources of Investment Information")
plt.show()
```

_The pie chart reveals that Financial Consultants (40%) and Newspapers and Magazines (35%) are the most relied-upon sources of investment information, indicating participants prefer professional advice and printed media. Television (15%) and the Internet (10%) play a smaller role_
